```python
def get_bb(x):
    return get_coords_BB(x["position"][0], x["position"][1], np.rad2deg(x["yaw"]), x["extent"][0], x["extent"][1])

def get_coords_BB(x, y, angle, extent_x, extent_y):
    endx1 = x - extent_x * math.sin(math.radians(angle)) - extent_y * math.cos(math.radians(angle))
    endy1 = y + extent_x * math.cos(math.radians(angle)) - extent_y * math.sin(math.radians(angle))

    endx2 = x + extent_x * math.sin(math.radians(angle)) - extent_y * math.cos(math.radians(angle))
    endy2 = y - extent_x * math.cos(math.radians(angle)) - extent_y * math.sin(math.radians(angle))

    endx3 = x + extent_x * math.sin(math.radians(angle)) + extent_y * math.cos(math.radians(angle))
    endy3 = y - extent_x * math.cos(math.radians(angle)) + extent_y * math.sin(math.radians(angle))

    endx4 = x - extent_x * math.sin(math.radians(angle)) + extent_y * math.cos(math.radians(angle))
    endy4 = y + extent_x * math.cos(math.radians(angle)) + extent_y * math.sin(math.radians(angle))

    return (endx1, endy1), (endx2, endy2), (endx3, endy3), (endx4, endy4)
```

```python
def viz_batch(batch, pix_per_m=5):
    batch_idxs, x_batch_objs, y_batch_objs, route_batch, waypoints_batch, targetpoint_batch = batch
    objs = x_batch_objs[batch_idxs[0]][1:]

    max_d = 30
    size = int(max_d*pix_per_m*2)
    origin = (size//2, size//2)
    PIXELS_PER_METER = pix_per_m

    img = np.zeros((size, size, 3), dtype=np.uint8)

    # route first
    for rp in batch[3][0]:
        x = rp[1]*PIXELS_PER_METER + origin[1] # TODO
        y = - rp[0]*PIXELS_PER_METER + origin[0]
        cv2.circle(img, (int(x), int(y)), 3, (255, 0, 0), -1)

    for o in objs:
        x = o[2]*PIXELS_PER_METER + origin[1]
        y = -o[1]*PIXELS_PER_METER + origin[0]
        yaw = o[3]
        extent_x = o[5]*PIXELS_PER_METER/2
        extent_y = o[6]*PIXELS_PER_METER/2
        origin_v = (x, y)
        vel = o[4]/3.6 # in m/s

        if o[0] in [1, 4]: # EGO and static car
            fill_color = [50,50,50]
            outline_color = [127,127,127]
        elif o[0] == 2: # Red for car
            fill_color = [0,0,100]
            outline_color = [0,0,255]
        elif o[0] == 3: # Cyan for walker
            fill_color = [100, 100, 0]
            outline_color = [255, 255, 0]
        elif o[0] == 5: # Red box for stop
            fill_color = [0, 0, 0]
            outline_color = [0, 0, 255]
        elif o[0] == 6:
            fill_color=[0,0,0] # Yellow box for traffic light
            outline_color = [0,255,255]
        else:
            fill_color = [127,127,127]
            outline_color = [255,0,255]

        if extent_x==0:
            print("X == 0")
            extent_x = 3
            extent_y = 3
        
        box = get_coords_BB(x, y, yaw-90, extent_x, extent_y)
        box = np.array(box).astype(int)
        draw = np.zeros_like(img)
        
        cv2.fillPoly(draw, [box], fill_color)
        cv2.drawContours(draw, [box], 0, outline_color, 1)

        # Creates speed indicator for vehicles and center dot for routes
        endx1, endy1, endx2, endy2 = get_coords(x, y, yaw-90, vel)
        cv2.line(draw, (int(endx1), int(endy1)), (int(endx2), int(endy2)), color=outline_color, thickness=1)

        img = cv2.addWeighted(img, 1, draw, 1, 0)

    for wp in batch[4][0]:
        x = wp[1]*PIXELS_PER_METER + origin[1] # TODO
        y = - wp[0]*PIXELS_PER_METER + origin[0]
        cv2.circle(img, (int(x), int(y)), 3, (0, 255, 0), -1)

    target_point = batch[5][0]
    x = target_point[1]*PIXELS_PER_METER + origin[1] # TODO
    y = - target_point[0]*PIXELS_PER_METER + origin[0]
    cv2.circle(img, (int(x), int(y)), 3, (255, 0, 200), -1)

    return img
```

```python
import torch
def generate_batch_v2(data_batch):
    maxseq = max([len(sample["input"]) for sample in data_batch])
    B = len(data_batch)

    batch_idxs = torch.zeros((B, maxseq))
    x_batch_objs = [[0,0,0,0,0,0,0]] # Placeholder
    y_batch_objs = [[0,0,0,0,0,0,0]] # Placeholder

    route_batch = []
    waypoints_batch = []
    targetpoint_batch = []
    light_bath = []

    n = 1
    for i, sample in enumerate(data_batch):
        #Input
        n_sample = len(sample["input"])
        batch_idxs[i, :n_sample] = torch.arange(n, n+n_sample)
        n += n_sample

        x_batch_objs.extend(sample["input"])
        y_batch_objs.extend(sample["output"]) # Assume that input and output is the same, idk if that is correct; aber fÃ¼r die preds sollte es doch gleich sein?; wie werden die sonst gematched
        route_batch.append(torch.tensor(sample["route"]))
        waypoints_batch.append(torch.tensor(sample["waypoints"]))
        targetpoint_batch.append(sample["target_point"])
        light_bath.append(sample["light"])

    x_batch_objs = torch.tensor(x_batch_objs) # B * max_seq x num_attributes + 1
    y_batch_objs = torch.tensor(y_batch_objs) # B * max_seq x num_attributes + 1
    route_batch = torch.stack(route_batch) # B x 15 x 2
    waypoints_batch = torch.stack(waypoints_batch) # B x 4 x 2
    targetpoint_batch = torch.tensor(targetpoint_batch) # B x 2
    light_bath = torch.tensor(light_bath)[..., None] # B x 1

    return batch_idxs, x_batch_objs, y_batch_objs, route_batch, waypoints_batch, targetpoint_batch, light_bath
```
